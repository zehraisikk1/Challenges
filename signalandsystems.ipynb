```python
import numpy as np
import matplotlib.pyplot as plt

# --- Signal Definitions ---

def rectangular_pulse(n, start_time, end_time, amplitude=1):
    """Generates a rectangular pulse."""
    signal = np.where((n >= start_time) & (n <= end_time), amplitude, 0)
    return signal

def triangular_signal(n, peak_time, width, amplitude=1):
    """Generates a triangular signal centered around peak_time."""
    signal = np.zeros_like(n, dtype=float)
    half_width = width // 2
    # Rising edge
    rising_mask = (n >= peak_time - half_width) & (n <= peak_time)
    if np.any(rising_mask): # Ensure there are points in the rising edge
        signal[rising_mask] = amplitude * (1 - (peak_time - n[rising_mask]) / half_width)
    # Falling edge
    falling_mask = (n > peak_time) & (n <= peak_time + half_width)
    if np.any(falling_mask): # Ensure there are points in the falling edge
        signal[falling_mask] = amplitude * (1 - (n[falling_mask] - peak_time) / half_width)
    return signal


def exponential_decay(n, decay_rate=0.8, amplitude=1, start_time=0):
    """Generates an exponentially decaying sequence starting from n=start_time."""
    signal = np.zeros_like(n, dtype=float)
    mask = n >= start_time
    signal[mask] = amplitude * (decay_rate ** (n[mask] - start_time))
    return signal

# --- Transformation Implementations ---

def time_reversal(x, n):
    """Performs time-reversal: y[n] = x[-n]."""
    # For visualization, we need to map the original n values to their new positions.
    # If original n is [-2, -1, 0, 1, 2] and x is [a, b, c, d, e]
    # x[-n] means:
    # y[-2] = x[2] = e
    # y[-1] = x[1] = d
    # y[0]  = x[0] = c
    # y[1]  = x[-1]= b
    # y[2]  = x[-2]= a
    # The new n_transformed will be -n_original[::-1] or sorted(n_original)
    # The signal y will be x_original[::-1]
    return x[::-1], n # The n vector for the new signal is also reversed in terms of values it represents

def time_shift(x, n, n0):
    """Performs time-shifting: y[n] = x[n-n0].
       n0 > 0: right shift (delay)
       n0 < 0: left shift (advance)
    """
    # The values of the signal x remain the same, but their corresponding n indices change.
    # If x[n] is defined for n, then x[n-n0] means the event that happened at n_orig now happens at n_orig + n0.
    # So, the new n vector is n_original + n0.
    n_shifted = n + n0
    return x, n_shifted

def fold_and_shift(x, n, n0):
    """Performs y[n] = x[-n+n0]."""
    # Step 1: Fold x[-n]
    x_folded = x[::-1]
    n_folded = -n[::-1] # The actual values of n for the folded signal
    # Step 2: Shift x_folded by n0 (to the right if n0 > 0)
    # y[n] = x_folded[n-n0]
    # This means the n values for x_folded are shifted by n0
    n_transformed = n_folded + n0
    return x_folded, n_transformed

def fold_and_shift_neg_n0(x, n, n0):
    """Performs y[n] = x[-n-n0]."""
    # Step 1: Fold x[-n]
    x_folded = x[::-1]
    n_folded = -n[::-1] # The actual values of n for the folded signal
    # Step 2: Shift x_folded by -n0 (to the left if n0 > 0, i.e., effectively x_folded[n - (-n0)])
    # y[n] = x_folded[n-(-n0)] = x_folded[n+n0]
    # This means the n values for x_folded are shifted by -n0
    n_transformed = n_folded - n0
    return x_folded, n_transformed


# --- Visualization Function ---
def plot_signals(n_orig, x_orig, n_transformed, x_transformed, title_orig, title_transformed, operation_description):
    """Plots original and transformed signals using stem plots."""
    fig, axs = plt.subplots(2, 1, figsize=(10, 6))

    # Original Signal
    axs[0].stem(n_orig, x_orig, linefmt='b-', markerfmt='bo', basefmt='r-')
    axs[0].set_title(title_orig)
    axs[0].set_xlabel('n')
    axs[0].set_ylabel('x[n]')
    axs[0].grid(True)
    axs[0].axvline(0, color='k', linestyle='--', linewidth=0.8, label='Origin (n=0)')
    axs[0].legend(loc='upper right')
    # Set x-ticks to be integers
    min_n_orig = np.min(n_orig)
    max_n_orig = np.max(n_orig)
    axs[0].set_xticks(np.arange(min_n_orig, max_n_orig + 1))


    # Transformed Signal
    axs[1].stem(n_transformed, x_transformed, linefmt='g-', markerfmt='go', basefmt='r-')
    axs[1].set_title(title_transformed)
    axs[1].set_xlabel('n')
    axs[1].set_ylabel('y[n]')
    axs[1].grid(True)
    axs[1].axvline(0, color='k', linestyle='--', linewidth=0.8, label='Origin (n=0)')
    axs[1].legend(loc='upper right')
    # Set x-ticks to be integers
    min_n_transformed = np.min(n_transformed)
    max_n_transformed = np.max(n_transformed)
    axs[1].set_xticks(np.arange(min_n_transformed, max_n_transformed + 1))


    fig.suptitle(f"Operation: {operation_description}", fontsize=14)
    plt.tight_layout(rect=[0, 0, 1, 0.96]) # Adjust layout to make space for suptitle
    plt.show()

# --- Mathematical Confirmation Function ---
def print_mathematical_confirmation(n_orig, x_orig, n_transformed, x_transformed, operation_name):
    """Prints numerical arrays before and after transformation."""
    print(f"\n--- Mathematical Confirmation for {operation_name} ---")
    print("Original Signal:")
    print(f"  n: {n_orig}")
    print(f"x[n]: {x_orig}")
    print("\nTransformed Signal:")
    # Sort by n for clearer comparison if n_transformed is not monotonic (e.g. after folding)
    sort_indices = np.argsort(n_transformed)
    print(f"  n': {n_transformed[sort_indices]}")
    print(f"y[n']: {x_transformed[sort_indices]}")
    print("--------------------------------------------------")

# --- Main Execution ---

# Define a common time vector for all signals
n_base = np.arange(-5, 11) # Extended range to see shifts clearly

# Signal Parameters
# Rectangular Pulse
rect_start = -1
rect_end = 3
rect_amp = 2
x_rect = rectangular_pulse(n_base, rect_start, rect_end, rect_amp)

# Triangular Signal
tri_peak = 2
tri_width = 6 # Must be an even number for this implementation
tri_amp = 3
x_tri = triangular_signal(n_base, tri_peak, tri_width, tri_amp)

# Exponentially Decaying Sequence
exp_decay_rate = 0.7
exp_amp = 4
exp_start_time = 0
x_exp = exponential_decay(n_base, exp_decay_rate, exp_amp, exp_start_time)


# List of signals and their names for iteration
signals = {
    "Rectangular Pulse": (n_base, x_rect),
    "Triangular Signal": (n_base, x_tri),
    "Exponential Decay": (n_base, x_exp)
}

# Shift amounts
n0_delay = 3  # For right shift x[n-n0]
n0_advance = 2 # For left shift x[n+n0] (use as -n0_advance in function call if n0 is positive for advance)
n0_fold_shift_pos = 2 # For x[-n+n0]
n0_fold_shift_neg = 2 # For x[-n-n0]


for sig_name, (n_orig, x_orig) in signals.items():
    print(f"\n\n======= Processing Signal: {sig_name} =======")

    # 1. Time-Reversal: y[n] = x[-n]
    x_folded, n_folded_display = time_reversal(np.copy(x_orig), np.copy(n_orig))
    # For time_reversal, the n_folded_display is -n_orig[::-1].
    # The actual values in x_folded correspond to these new n indices.
    # To plot correctly, we need to ensure the n vector for plotting corresponds to the values.
    # The x_folded is x_orig[::-1]. The n for these values should be -n_orig[::-1].
    # If n_orig = [-2, -1, 0, 1, 2], x_orig = [a, b, c, d, e]
    # x_folded = [e, d, c, b, a]
    # n_for_folded_x = [-2, -1, 0, 1, 2] after folding, these points correspond to x at indices [2, 1, 0, -1, -2] of original n
    # So, n values are effectively -n_orig in reverse order.
    n_plot_folded = -n_orig[::-1] # These are the n values where the folded signal samples will appear.

    plot_signals(n_orig, x_orig, n_plot_folded, x_folded,
                 f"Original {sig_name}: x[n]",
                 f"Folded {sig_name}: y[n]=x[-n]",
                 f"Time-Reversal (Folding) for {sig_name}")
    print_mathematical_confirmation(n_orig, x_orig, n_plot_folded, x_folded, f"Time-Reversal (Folding) - {sig_name}")

    # 2. Right Shift (Delay): y[n] = x[n-n0]
    x_delayed, n_delayed = time_shift(np.copy(x_orig), np.copy(n_orig), n0_delay)
    plot_signals(n_orig, x_orig, n_delayed, x_delayed,
                 f"Original {sig_name}: x[n]",
                 f"Delayed {sig_name}: y[n]=x[n-{n0_delay}] (Right Shift by {n0_delay})",
                 f"Time-Shifting (Delay) for {sig_name}")
    print_mathematical_confirmation(n_orig, x_orig, n_delayed, x_delayed, f"Right Shift (Delay by {n0_delay}) - {sig_name}")

    # 3. Left Shift (Advance): y[n] = x[n+n0]
    # Our time_shift function takes n0, where positive n0 is a delay.
    # So for x[n+n0_advance], we use -n0_advance as the shift amount.
    x_advanced, n_advanced = time_shift(np.copy(x_orig), np.copy(n_orig), -n0_advance)
    plot_signals(n_orig, x_orig, n_advanced, x_advanced,
                 f"Original {sig_name}: x[n]",
                 f"Advanced {sig_name}: y[n]=x[n+{n0_advance}] (Left Shift by {n0_advance})",
                 f"Time-Shifting (Advance) for {sig_name}")
    print_mathematical_confirmation(n_orig, x_orig, n_advanced, x_advanced, f"Left Shift (Advance by {n0_advance}) - {sig_name}")

    # 4. Combination: y[n] = x[-n+n0]
    x_fold_shift_pos, n_fold_shift_pos = fold_and_shift(np.copy(x_orig), np.copy(n_orig), n0_fold_shift_pos)
    plot_signals(n_orig, x_orig, n_fold_shift_pos, x_fold_shift_pos,
                 f"Original {sig_name}: x[n]",
                 f"Transformed {sig_name}: y[n]=x[-n+{n0_fold_shift_pos}]",
                 f"Fold and Shift (x[-n+{n0_fold_shift_pos}]) for {sig_name}")
    print_mathematical_confirmation(n_orig, x_orig, n_fold_shift_pos, x_fold_shift_pos, f"Fold and Shift (x[-n+{n0_fold_shift_pos}]) - {sig_name}")

    # 5. Combination: y[n] = x[-n-n0]
    x_fold_shift_neg, n_fold_shift_neg = fold_and_shift_neg_n0(np.copy(x_orig), np.copy(n_orig), n0_fold_shift_neg)
    plot_signals(n_orig, x_orig, n_fold_shift_neg, x_fold_shift_neg,
                 f"Original {sig_name}: x[n]",
                 f"Transformed {sig_name}: y[n]=x[-n-{n0_fold_shift_neg}]",
                 f"Fold and Shift (x[-n-{n0_fold_shift_neg}]) for {sig_name}")
    print_mathematical_confirmation(n_orig, x_orig, n_fold_shift_neg, x_fold_shift_neg, f"Fold and Shift (x[-n-{n0_fold_shift_neg}]) - {sig_name}")

print("\nAll operations complete.")
```
